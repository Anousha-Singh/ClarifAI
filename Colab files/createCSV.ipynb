```python
import glob
import pandas as pd
import os
```

```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Define paths
base_path = '/content/drive/My Drive/preprocessedDataset/'
fake_folder = os.path.join(base_path, 'fake')
real_folder = os.path.join(base_path, 'real')
output_csv_path = os.path.join(base_path, 'videoLabels.csv')

# Create lists to store file information
all_files = []
labels = []
```

```python
# Process fake files
print("Processing 'fake' folder...")
fake_files = glob.glob(os.path.join(fake_folder, '*.mp4'))
for file_path in fake_files:
    filename = os.path.basename(file_path)
    all_files.append(filename)
    labels.append('fake')
print(f"Found {len(fake_files)} MP4 files in the 'fake' folder")
```

```python
# Process real files
print("Processing 'real' folder...")
real_files = glob.glob(os.path.join(real_folder, '*.mp4'))
for file_path in real_files:
    filename = os.path.basename(file_path)
    all_files.append(filename)
    labels.append('real')
print(f"Found {len(real_files)} MP4 files in the 'real' folder")
```

```python
# Create DataFrame
df = pd.DataFrame({
    'filename': all_files,
    'class': labels
})

# Save to CSV
df.to_csv(output_csv_path, index=False)
print(f"CSV file created at: {output_csv_path}")
print(f"Total files processed: {len(all_files)}")

# Display the first few rows of the DataFrame
print("\nSample of the created CSV:")
print(df.head())
```
