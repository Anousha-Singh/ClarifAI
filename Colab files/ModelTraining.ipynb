```python
import torch
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import face_recognition

class VideoDataset(Dataset):
    def __init__(self, video_names, labels, sequence_length=20, transform=None):
        """
        Dataset for loading video data.

        Args:
            video_names (list): List of video file paths
            labels (DataFrame): DataFrame containing file names and labels
            sequence_length (int): Number of frames to sample from each video
            transform (callable): Transformations to apply to frames
        """
        print(len(video_names))

        # Filter out non-existent or corrupted videos first
        self.video_names = []

        for video_path in video_names:
            if os.path.exists(video_path) and self._check_video(video_path, sequence_length):   #Check if path exists and the vidoe is not corrupted
                self.video_names.append(video_path)

        print(f"Valid videos for dataset: {len(self.video_names)}/{len(video_names)}")          #Valid/Total
        # print(self.video_names)

        self.labels = labels
        self.transform = transform
        self.count = sequence_length



    def _check_video(self, video_path, min_frames=10):
        """Check if video file exists and has enough frames"""
        try:
            cap = cv2.VideoCapture(video_path)
            if not cap.isOpened():
                print(f"Failed to open video: {video_path}")
                return False

            frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            cap.release()

            if frame_count < min_frames:
                print(f"Video too short ({frame_count} frames): {video_path}")
                return False

            return True
        except Exception as e:
            print(f"Error checking video {video_path}: {e}")
            return False

    def __len__(self):
        return len(self.video_names)


    #Special method in python, we can access a frame of video though index
    def __getitem__(self, idx):

        if idx >= len(self.video_names):
            raise IndexError(f"Index {idx} out of range (max {len(self.video_names)-1})")

        video_path = self.video_names[idx]
        frames = []

        # Get video filename without path
        temp_video = os.path.basename(video_path)

        try:
            # Find the label for this video
            label_indices = self.labels.loc[self.labels["file"] == temp_video].index.values
            if len(label_indices) == 0:
                print(f"Warning: No label found for {temp_video}")
                # Default to 'FAKE' if no label found
                label = 0
            else:
                label_text = self.labels.iloc[label_indices[0], 1]
                label = 0 if label_text == 'fake' else 1

            # Extract frames
            cap = cv2.VideoCapture(video_path)
            total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

            if total_frames <= 0:
                raise ValueError(f"Video has no frames: {video_path}")

            # Calculate frame indices to sample evenly through the video
            indices = np.linspace(0, total_frames - 1, self.count, dtype=int)

            for i in indices:
                cap.set(cv2.CAP_PROP_POS_FRAMES, i)
                ret, frame = cap.read()
                if not ret:
                    raise ValueError(f"Failed to read frame {i} from {video_path}")

                #Proccess the frame and transform it in desirable size
                if self.transform:
                    frame = self.transform(frame)
                frames.append(frame)

            cap.release()

            # Ensure we have the required number of frames
            if len(frames) < self.count:
                raise ValueError(f"Could only extract {len(frames)}/{self.count} frames from {video_path}")

            # Stack frames into a single tensor
            frames = torch.stack(frames)

            return frames, label

        except Exception as e:
            print(f"Error processing video {video_path}: {e}")
            # Return a dummy tensor in case of failure to prevent training loop from breaking
            # Create a dummy tensor with the right dimensions
            dummy_frames = torch.zeros((self.count, 3, 112, 112))
            return dummy_frames, 0  # Return as fake (0)

    def frame_extract(self, path):
          """Generator to extract frames from a video"""
          vidObj = cv2.VideoCapture(path)

          if not vidObj.isOpened():
              print(f"Failed to open video: {path}")
              return

          frame_count = int(vidObj.get(cv2.CAP_PROP_FRAME_COUNT))
          print(f"Extracting frames from: {path} (Total frames: {frame_count})")

          for _ in range(frame_count):
              success, image = vidObj.read()
              if not success:
                  print(f"Warning: Unable to read frame at index {_} in {path}")
                  break
              yield image

          vidObj.release()



# Function to display a frame
# From preprocessed image to original image
def im_plot(tensor):
    """Plot an image tensor"""
    image = tensor.cpu().numpy().transpose(1, 2, 0)
    b, g, r = cv2.split(image)
    image = cv2.merge((r, g, b))
    # Denormalize
    image = image * np.array(std) + np.array(mean)
    image = np.clip(image * 255.0, 0, 255)
    plt.imshow(image.astype(np.uint8))
    plt.show()
```

```python
def number_of_real_and_fake_videos(data_list):
    real = 0
    fake = 0
    for i in data_list:
        temp_video = os.path.basename(i)
        label = video_labels.get(temp_video, 'fake')  # Default to 'fake' if label not found
        if label == 'fake':
            fake += 1
        elif label == 'real':
            real += 1
    return real, fake
```

```python
# load the labels and video in data loader
import pandas as pd
from collections import Counter
from sklearn.model_selection import train_test_split

header_list = ["file","label"]
labels = pd.read_csv('/content/drive/My Drive/preprocessedDataset/videoLabels.csv',names=header_list)

############

# Filters out label that are not real or fake
labels = labels[labels['label'].isin(['real', 'fake'])]

# Recreate dictionary after filtering
video_labels = labels.set_index('file')['label'].to_dict()

# Label list
video_labels_list = list(video_labels.values())

# Create a list of video files based on the loaded labels
video_files = glob.glob(os.path.join(fake_folder, "*.mp4")) + glob.glob(os.path.join(real_folder, "*.mp4"))


# Create a list of labels for the video files
video_labels_list = [video_labels[os.path.basename(v)] for v in video_files]

# Verify that video_labels_list and video_files have the same length
assert len(video_files) == len(video_labels_list), "Mismatch between video files and labels."


# Count occurrences of each label
label_counts = Counter(video_labels_list)
print("Label distribution:", label_counts)

# Perform stratified train-test split
# Only apply stratify if both classes have at least two samples
if min(label_counts.values()) >= 2:
    train_videos, valid_videos = train_test_split(video_files, test_size=0.2, stratify=video_labels_list)
else:
    train_videos, valid_videos = train_test_split(video_files, test_size=0.2, stratify=None)  # No stratification


train_real, train_fake = number_of_real_and_fake_videos(train_videos)
valid_real, valid_fake = number_of_real_and_fake_videos(valid_videos)

print("TRAIN: Real:", train_real, " Fake:", train_fake)
print("TEST: Real:", valid_real, " Fake:", valid_fake)

###########


im_size = 112
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

train_transforms = transforms.Compose([
                                        transforms.ToPILImage(),
                                        transforms.Resize((im_size,im_size)),
                                        transforms.ToTensor(),
                                        transforms.Normalize(mean,std)])

test_transforms = transforms.Compose([
                                        transforms.ToPILImage(),
                                        transforms.Resize((im_size,im_size)),
                                        transforms.ToTensor(),
                                        transforms.Normalize(mean,std)])
train_data = VideoDataset(train_videos,labels,sequence_length = 10,transform = train_transforms)
#print(train_data)
val_data = VideoDataset(valid_videos,labels,sequence_length = 10,transform = train_transforms)
train_loader = DataLoader(train_data,batch_size = 4,shuffle = True,num_workers = 2)
valid_loader = DataLoader(val_data,batch_size = 4,shuffle = True,num_workers = 2)
image,label = train_data[0]
im_plot(image[0,:,:,:])
```

```python
from torch import nn
from torchvision.models import resnext50_32x4d, ResNeXt50_32X4D_Weights

class Model(nn.Module):
    def __init__(self, num_classes, latent_dim=2048, lstm_layers=1, hidden_dim=2048, bidirectional=False):
        super(Model, self).__init__()

        weights = ResNeXt50_32X4D_Weights.DEFAULT # For feature extraction
        backbone = resnext50_32x4d(weights=weights)  # Load pretrained ResNeXt50

        self.model = nn.Sequential(*list(backbone.children())[:-2])  # Remove final Final Classification layers
        self.avgpool = nn.AdaptiveAvgPool2d(1)                   # Reduce feature map size

        self.lstm = nn.LSTM(latent_dim, hidden_dim, lstm_layers, bidirectional)  # For sequential modeling
        self.relu = nn.LeakyReLU()          #####
        self.dp = nn.Dropout(0.4)    # Randomly disables 40% of neurons during training to prevent overfitting.

        self.linear1 = nn.Linear(hidden_dim, num_classes)  #For binary classification

    def forward(self, x):
        batch_size, seq_length, c, h, w = x.shape   #(Batch, Time, Channels, Height, Width)
        x = x.view(batch_size * seq_length, c, h, w)

        fmap = self.model(x)  # CNN features
        x = self.avgpool(fmap)  # (B*T, 2048, 1, 1)
        x = x.view(batch_size, seq_length, -1)  # reshape for LSTM: (B, T, 2048)

        x_lstm, _ = self.lstm(x, None)
        x_mean = torch.mean(x_lstm, dim=1)  # temporal average
        x_out = self.dp(self.linear1(x_mean))  # classification

        return fmap, x_out
```

```python
import torch
from torch.autograd import Variable
import time
import os
import sys
import os

def train_epoch(epoch, num_epochs, data_loader, model, criterion, optimizer):
    model.train()
    losses = AverageMeter()
    accuracies = AverageMeter()

    for i, (inputs, targets) in enumerate(data_loader):
        # If GPU available
        if torch.cuda.is_available():
            targets = targets.type(torch.cuda.LongTensor)
            inputs = inputs.cuda()

        _,outputs = model(inputs)

        # How far off the predictions are from the true labels
        loss  = criterion(outputs,targets.type(torch.cuda.LongTensor))

        # Accuracy
        acc = calculate_accuracy(outputs, targets.type(torch.cuda.LongTensor))
        losses.update(loss.item(), inputs.size(0))
        accuracies.update(acc, inputs.size(0))

        # Backpropogation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()  # Adjust weights

        sys.stdout.write(
                "\r[Epoch %d/%d] [Batch %d / %d] [Loss: %f, Acc: %.2f%%]"
                % (
                    epoch,
                    num_epochs,
                    i,
                    len(data_loader),
                    losses.avg,
                    accuracies.avg))
    # Save model
    torch.save(model.state_dict(),'/content/drive/My Drive/model.pt')
    return losses.avg,accuracies.avg

def test(epoch, model, data_loader, criterion):
    print('Testing')
    model.eval()
    losses = AverageMeter()
    accuracies = AverageMeter()
    pred = []
    true = []

    with torch.no_grad():
        for i, (inputs, targets) in enumerate(data_loader):
            inputs = inputs.to(next(model.parameters()).device)
            targets = targets.long().to(inputs.device)

            _, outputs = model(inputs)

            # Sanity check for bad outputs
            if torch.isnan(outputs).any() or torch.isinf(outputs).any():
                print("NaN or Inf detected in model outputs!")
                continue

            loss = criterion(outputs, targets)
            acc = calculate_accuracy(outputs, targets)

            _, p = torch.max(outputs, 1)
            true += targets.cpu().numpy().tolist()
            pred += p.cpu().numpy().tolist()

            losses.update(loss.item(), inputs.size(0))
            accuracies.update(acc, inputs.size(0))

            sys.stdout.write(
                "\r[Batch %d / %d]  [Loss: %f, Acc: %.2f%%]"
                % (
                    i,
                    len(data_loader),
                    losses.avg,
                    accuracies.avg,
                )
            )

        print('\nAccuracy {:.2f}%'.format(accuracies.avg))

    return true, pred, losses.avg, accuracies.avg

class AverageMeter(object):
    """Computes and stores the average and current value"""
    def __init__(self):
        self.reset()
    def reset(self):
        self.val = 0
        self.avg = 0
        self.sum = 0
        self.count = 0

    def update(self, val, n=1):
        self.val = val
        self.sum += val * n  # Add current value * number of samples in a single batch
        self.count += n
        self.avg = self.sum / self.count

def calculate_accuracy(outputs, targets):
    batch_size = targets.size(0)
    _, pred = outputs.topk(1, 1, True)
    pred = pred.t()
    correct = pred.eq(targets.view(1, -1))
    n_correct_elems = correct.float().sum().item()
    return 100* n_correct_elems / batch_size
```

```python
def plot_loss(train_loss_avg,test_loss_avg,num_epochs):
  loss_train = train_loss_avg
  loss_val = test_loss_avg
  print(num_epochs)
  epochs = range(1,num_epochs+1)
  plt.plot(epochs, loss_train, 'g', label='Training loss')
  plt.plot(epochs, loss_val, 'b', label='validation loss')
  plt.title('Training and Validation loss')
  plt.xlabel('Epochs')
  plt.ylabel('Loss')
  plt.legend()
  plt.show()
def plot_accuracy(train_accuracy,test_accuracy,num_epochs):
  loss_train = train_accuracy
  loss_val = test_accuracy
  epochs = range(1,num_epochs+1)
  plt.plot(epochs, loss_train, 'g', label='Training accuracy')
  plt.plot(epochs, loss_val, 'b', label='validation accuracy')
  plt.title('Training and Validation accuracy')
  plt.xlabel('Epochs')
  plt.ylabel('Accuracy')
  plt.legend()
  plt.show()
```
